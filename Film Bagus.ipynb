```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

## Content Based Filtering

```python
dfAnime = pd.read_csv('movies.csv')
dfAnime
```

```python
dfAnime = dfAnime[dfAnime['genres'].notna()]

judul = 'Superman vs. The Elite (2012)'
cari = dfAnime.iloc[dfAnime['title'].tolist().index(judul)]['genres'].replace(' ','')
cari
```

```python
CV = CountVectorizer()
dfSkor = pd.DataFrame()
for i in range(len(dfAnime['genres'])):
    cm = CV.fit_transform([cari, dfAnime.iloc[i]['genres']])
    similarityScore = cosine_similarity(cm.toarray())
    similarityScore = pd.DataFrame(similarityScore)
    score = similarityScore[0].sort_values(ascending=False)
    dfSkor = dfSkor.append([score[1]])
dfSkor = dfSkor.reset_index()
dfSkor = dfSkor[0].sort_values(ascending=False)
```

```python
top_5_indexes = list(dfSkor.iloc[0:6].index)
recommended_anime=[]
for i in top_5_indexes:
        if (dfAnime.iloc[i]['title']) == judul:
            continue
        recommended_anime.append(dfAnime.iloc[i]['title'])
        print('Movie ID =',dfAnime.iloc[i]['movieId'],'| Title =', dfAnime.iloc[i]['title'])
```

## Collaborative Filtering

```python
dfRating = pd.read_csv('Ratings.csv')
```

```python
dfRating.head(3)
```

```python
# membuang column timestamp

dfRating.drop(columns =['timestamp'], inplace=True)
dfRating
```

```python
dfRating2 = dfRating.pivot_table(
    index = 'userId',
    columns = 'movieId'
)
dfRating2
```

```python
dfRating2 = dfRating2.fillna(0)
dfRating2
```

```python
dfRating2cor = dfRating2.corr().loc['rating']
dfRating2cor
```

```python
widodo = [('Being Flynn (2012)', 5)]
indexWidodo = dfAnime['title'].tolist().index(widodo[0][0])
favWidodo = [(dfAnime.iloc[indexWidodo]['movieId'], widodo[0][1])]
favWidodo
```

```python
dfRating2cor.loc[95816]
```

```python
scoreSimilar = pd.DataFrame()
for anime_id, rating in favWidodo:
    score = dfRating2cor.loc[anime_id] * (rating/5)
    score = score.sort_values(ascending=False)
    scoreSimilar = scoreSimilar.append(score, ignore_index=True)
scoreSimilar
```

```python
dfFinal = pd.DataFrame(scoreSimilar.sum().sort_values(ascending=False))
```

```python
dfFinal[0:6]
```

```python
i = 0
print('Rekomendasi untuk Widodo')
for rating, idMovie in dfFinal.index[0:6]:
    if idMovie == favWidodo[0][0]:
        continue
    if i == 5:
        continue
    print('Movie ID = ', idMovie,'| Title =', dfAnime.iloc[dfAnime['movieId'].tolist().index(idMovie)]['title'])
    i+=1
```
